```python
#importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#READING THE DATA

MYPATH = "/Users/redhak/Downloads/SampleSuperstore.csv"
store = pd.read_csv(MYPATH)
print("Data imported successfully")

store.head(10)
```

```python
plt.figure(figsize=(10,15))
sns.heatmap(store.corr(),annot=True)
plt.show()
```

From heatmap, we see positive correlation between Sales and Profit and negative correlation between Profit and Discount

```python
#trend of profit wrt discount values
plt.figure(figsize=(10,15))
sns.lineplot(x= 'Discount',y= 'Profit',data=store)
plt.show()
```

```python
#statewise profit
profit_sum = store.groupby("State")["Profit"].sum()
profit_sum
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.xlabel("States")
plt.ylabel("Profits")
plt.xticks(rotation=90)
plt.show()
```

Highest Profit state is California and lowest Profit state is Texas

```python
sales_sum= store.groupby("State")["Sales"].sum()
sales_sum
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x=sales_sum.index, y=sales_sum.values)
plt.xlabel("States")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```


Highest Sales state is Calfornia and lowest Sales state is North Dakota

```python
#Comparing Profits and Sales of the states
profits_sales = store.groupby("State")["Profit","Sales"].sum()
plt.figure(figsize=(20,15))
sns.barplot(x=profit_sum,y=sales_sum)
plt.xticks(rotation=90)
plt.show()
```

```python
#Comparing Profits and Sales of the states
profits_sales = store.groupby("State")["Profit","Sales"].sum()
profits_sales.plot.bar(figsize=(20,10))
plt.show()
```

California has highest sales and highest profit. Texas, Ohio,Illinois and Pennsylvania have high sales but are still in loss

```python
#Average discount percent of each state
avg_discount_percent = store.groupby("State")["Discount"].mean()*100
avg_discount_percent
```

```python
#plotting avg discount percent of each state
plt.figure(figsize=(20,10))
sns.barplot(x=avg_discount_percent.index,y=avg_discount_percent.values)
plt.ylabel("Avg Discount Percent")
plt.xticks(rotation=90)
plt.show()
```

Illinois has highest Average discount percent

```python
plt.figure(figsize=(20,15))
sns.barplot(data= store, x="Region", y="Profit",hue="Segment")
plt.xticks(rotation=90)
plt.show()
```

Max profits through Home Office is in East Region and min is in South Region.

```python
category_data = store.groupby(['Category'])[['Sales','Discount','Profit']].mean()
category_data
```

```python
#pie chart of category wise sales , profits and discount
category_data.plot.pie(subplots=True, figsize=(20,15), labels = category_data.index)
plt.show()
```

We can see Technology category has maximum Sales & Profit. Office supplies category has minimum Sales and Furniture category has lowest Profit.

```python
shipmode_data = store.groupby(["Ship Mode"])[["Sales", "Discount", "Profit"]].mean()
shipmode_data
```

```python
shipmode_data.plot.pie(subplots=True, figsize=(20,15),labels=shipmode_data.index)
plt.show()
```

Same Day Ship Mode has highest sales. First Class Ship Mode has highest profit and discount. Standard Ship Mode has lowest Sales and Profit

```python
subcategory_data = store.groupby(["Sub-Category"])[["Profit","Sales"]].mean()
subcategory_data
```

```python
subcategory_data.plot.bar(figsize=(20,15))
plt.show()
```

We can see that Copiers have highest Sales and Profit, whereas Tables have lowest Profit and Fasteners have lowest Sales.
